```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC, SVR
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, r2_score, mean_absolute_error, mean_squared_error
```

```python
rdf = pd.read_csv('Student_Performance.csv')
rdf
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
rdf['Extracurricular Activities'] = encoder.fit_transform(rdf['Extracurricular Activities'])
rdf
```

```python
x = rdf.drop('Performance Index', axis=1)
y = rdf['Performance Index']
```

```python
y.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
rmodel = SVR(kernel='linear', C=10, epsilon=0.01)
rmodel.fit(xtrain, ytrain)
```

```python
ypred = rmodel.predict(xtest)
ypred
```

```python
print('Training Efficiency =', rmodel.score(xtrain, ytrain))
print('Testing Efficiency =', rmodel.score(xtest, ytest))
```
