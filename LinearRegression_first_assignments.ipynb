```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://github.com/Jeevita-R/ML_datasets/raw/refs/heads/main/House_price_prediction.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
label = LabelEncoder()
df['Type_encoded'] = label.fit_transform(df['Type'])
df
```

```python
df['Method_encoded'] = label.fit_transform(df['Method'])
df
```

#pd.get_dummies(df,dtypes = int)
applies one hot encoding in whole datasets in one time

```python
df['Region_encoded'] = label.fit_transform(df['Regionname'])
df
```

```python
df.drop(columns = ['Type', 'Method', 'Regionname'], axis = 1, inplace = True)
```

```python
df
```

```python
onehot = pd.get_dummies(df, columns = ['Suburb', 'SellerG', 'CouncilArea'], dtype = int)
onehot
```

```python
df
```

```python
onehot.shape
```

```python
x = onehot.drop(columns = ['Price'], axis = 1)
y = onehot['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
n = x_test.shape[0]
p = x_test.shape[1]
adj_r2 = 1 - ((1-r2)*(n-1)/(n-p-1))
adj_r2
```
