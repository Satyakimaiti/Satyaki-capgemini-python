```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
import seaborn as sns
```

```python
data = pd.read_csv('Melbourne.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.hist(figsize=(15,10))
plt.show()
```

```python
data.skew(numeric_only=True)
```

```python
for c in data.columns:
    if data[c].dtype != 'object':
        s = data[c].skew()
        
        if s > 0.5:
            data[c].fillna(data[c].median(), inplace=True)
        else:
            data[c].fillna(data[c].mean(), inplace=True)
    else:
        data[c].fillna(data[c].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
data.hist(figsize=(15,10))
plt.show()
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data = data.drop(columns=['Address','Date'], axis = 1)
```

```python
data.head()
```

```python
data['Suburb'].nunique()
```

```python
le = LabelEncoder()
data['Type'] = le.fit_transform(data['Type'])
```

```python
data = pd.get_dummies(data,columns=['Suburb', 'Method', 'SellerG', 'CouncilArea', 'Regionname'],dtype=int)
```

```python
data.drop(columns=['Suburb', 'Method', 'SellerG', 'CouncilArea', 'Regionname'], axis=1, errors='ignore', inplace=True)
```

```python
data
```

```python
scaler = StandardScaler()
data[['Postcode','Price','Landsize','BuildingArea']] = scaler.fit_transform(data[['Postcode','Price','Landsize','BuildingArea']])
```

```python
data
```

```python
X = data.drop(columns=['Price'])
y = data['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
model.score(X_train, y_train)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
ridge = Ridge(alpha=50)
ridge.fit(X_train, y_train)
y_pred = ridge.predict(X_test)
ridge.score(X_test, y_test)
```

```python
ridge.score(X_train, y_train)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
lasso = Lasso(alpha=55)
lasso.fit(X_train, y_train)
y_pred = lasso.predict(X_test)
lasso.score(X_test, y_test)
```

```python
lasso.score(X_train, y_train)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2 = r2_score(y_test,y_pred)
r2
```
