```python
import pandas as pd
from sklearn.datasets import load_wine

wine = load_wine(as_frame = True)
df = wine.frame

print(df.head())
```

```python
X = df.drop('target', axis = 1)
Y = df['target']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_standard = scaler.fit_transform(X)

X_standard_df = pd.DataFrame(X_standard, columns = X.columns)

print(X_standard_df.head())
```

```python
data = pd.DataFrame({
    'Sname': ['S1', 'S2', 'S3', 'S4'],
    'Height': [162, 171, 155, 165],
    'Weight': [63, 60, 50, 62],
    'Speed': [5, 3, 7, 2]
})
print(data.head())
```

```python
from sklearn.preprocessing import Normalizer

norm = Normalizer()

num_data = data[['Height', 'Weight', 'Speed']]
data_norm = norm.fit_transform(num_data)
df = pd.DataFrame(data_norm, columns = num_data.columns)
df.head()
```
