```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Measurement_info.csv', encoding = 'ISO-8859-10')
data
```

```python
item = pd.read_csv('Measurement_item_info.csv', encoding = 'ISO-8859-10')
item
```

```python
station = pd.read_csv('Measurement_station_info.csv', encoding = 'ISO-8859-10')
station
```

```python
sub_item = item[['Item code', 'Item name']]
sub_item
```

```python
sub_station = station[['Station code', 'Station name(district)']]
sub_station
```

```python
data_i = data.merge(item[['Item code', 'Item name']],
                    on = 'Item code',
                    how = 'left')
data_i
```

```python
data_s = data_i.merge(station[['Station code', 'Station name(district)']],
                      on = 'Station code',
                      how = 'left')
data_s
```

```python
data = data_s.drop(['Item code', 'Station code'], axis = 1)
data
```

```python
status_dict = {'Instrument status': [0, 1, 2, 4, 8, 9],
 'Status': ['Normal',
  'Need for calibration',
  'Abnormal',
  'Power cut off',
  'Under repair',
  'Abnormal data']}

status_dict
```

```python
dictdf = pd.DataFrame(status_dict)
dictdf
```

```python
data
```

```python
data = data.loc[:, ~data.columns.duplicated()]
data
```

```python
data = data.merge(
    dictdf[['Instrument status', 'Status']],
    on='Instrument status',
    how='left'
)
data
```

```python
data['Year'] = pd.DatetimeIndex(data['Measurement date']).year
data['Month'] = pd.DatetimeIndex(data['Measurement date']).month
data['Day'] = pd.DatetimeIndex(data['Measurement date']).day
data['Hour'] = pd.DatetimeIndex(data['Measurement date']).hour
data.head()
```

```python
data = data.drop(["Measurement date"], axis =1)
```
