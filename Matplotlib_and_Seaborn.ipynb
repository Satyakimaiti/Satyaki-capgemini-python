```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('df_final_features.csv')
data = df.head(200)


cols = data.select_dtypes(include = 'number').columns

x = data[cols[0]]
y = data[cols[1]]
```

Line **Plot**

```python
plt.figure(figsize = (10, 5))

plt.title('Line Plot')

plt.plot(x, marker = 'o')
plt.xlabel(cols[0])
plt.ylabel('values')

#plt.xticks(rotation = 90)
plt.show()
```

Bar **Plot**

```python
plt.figure(figsize = (10,5))

plt.title('Bar Plot')
plt.bar(range(len(x)), x,
        color = 'orange',
        edgecolor = 'black')
plt.xlabel('index')
plt.ylabel(cols[0])

plt.legend()
plt.show()
```

**Histogram**

```python
plt.figure(figsize=(10,5))

plt.title('Histogram')

plt.hist(x,
         bins=10,
         color='skyblue',
         edgecolor='black')

plt.xlabel(cols[0])
plt.ylabel('Frequency')

plt.show()
```

Scatter **Plot**

```python
plt.figure(figsize=(10,5))

plt.title('Scatter Plot')

plt.scatter(x, y,
            color='green',
            marker='o')

plt.xlabel(cols[0])
plt.ylabel(cols[1])

plt.xticks(rotation=45)
plt.show()
```

Box **Plot**

```python
plt.figure(figsize=(10,3))

plt.title('Box Plot')

plt.boxplot(x)

plt.xlabel('Data')
plt.ylabel(cols[0])

plt.show()
```

Pie **Chart**

```python
top5 = x.value_counts().head(5)

plt.figure(figsize=(10,3))

plt.title('Pie Chart')

plt.pie(top5,
        labels=top5.index,
        autopct='%1.1f%%')

plt.show()
```

Area **Plot**

```python
plt.figure(figsize=(10,3))

plt.title('Area Plot')

plt.fill_between(range(len(x)), x,
                 color='lightgreen')

plt.xlabel('Index')
plt.ylabel(cols[0])

plt.xticks(rotation=45)
plt.show()
```

Stem **Plot**

```python
plt.figure(figsize=(10,5))

plt.title('Stem Plot')

plt.stem(x.head(50))

plt.xlabel('Index')
plt.ylabel(cols[0])

plt.show()
```

Step **Plot**

```python
plt.figure(figsize=(10,3))

plt.title('Step Plot')

plt.step(range(len(x)), x,
         color='brown',
         linestyle='--')

plt.xlabel('Index')
plt.ylabel(cols[0])

plt.xticks(rotation=45)
plt.show()
```

Seaborn **Code**

Line **Plot**

```python
import seaborn as sns
plt.figure(figsize=(10,3))

plt.title('Line Chart')

sns.lineplot(x=x, y=x,
             marker='o',
             linestyle='--')

plt.xlabel(cols[0])
plt.ylabel(cols[0])

plt.xticks(rotation=45)
plt.show()
```

Bar **Plot**

```python
plt.figure(figsize=(10,3))

plt.title('Bar Chart')

sns.barplot(x=x.index, y=x)

plt.xlabel('Index')
plt.ylabel(cols[0])

plt.xticks(rotation=45)
plt.show()
```

**Histogram**

```python
plt.figure(figsize=(10,3))

plt.title('Histogram')

sns.histplot(x=x, bins=10)

plt.xlabel(cols[0])
plt.ylabel('Frequency')

plt.show()
```

Scatter **Plot**

```python
plt.figure(figsize=(10,3))

plt.title('Scatter Plot')

sns.scatterplot(x=x, y=y)

plt.xlabel(cols[0])
plt.ylabel(cols[1])

plt.xticks(rotation=45)
plt.show()
```

Box **Plot**

```python
plt.figure(figsize=(10,3))

plt.title('Box Plot')

sns.boxplot(x=x)

plt.xlabel(cols[0])
plt.show()
```

Pie **Chart**

```python
top5 = x.value_counts().head(5)

plt.figure(figsize=(10,3))

plt.title('Pie Chart')

plt.pie(top5, labels=top5.index, autopct='%1.1f%%')

plt.show()
```

Area **Plot**

```python
plt.figure(figsize=(10,3))

plt.title('Area Plot')

sns.lineplot(x=x.index, y=x)
plt.fill_between(x.index, x, alpha=0.3)

plt.xlabel('Index')
plt.ylabel(cols[0])

plt.xticks(rotation=45)
plt.show()
```

Violin **Plot**

```python
plt.figure(figsize=(10,3))

plt.title('Violin Plot')

sns.violinplot(x=x)

plt.xlabel(cols[0])
plt.show()
```

**HeatMap**

```python
plt.figure(figsize=(10,3))

plt.title('Correlation Heatmap')

sns.heatmap(df[cols].corr(), annot=True)

plt.show()
```
