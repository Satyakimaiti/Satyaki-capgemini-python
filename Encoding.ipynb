```python
data = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'Education': ['Bachelors', 'Masters', 'PhD', 'Bachelors', 'Masters'],
    'City': ['Kolkata', 'Delhi', 'Mumbai', 'Delhi', 'Kolkata'],
    'Age': [23, 35, 45, 33, 28],
    'Salary': [30000, 60000, 80000, 50000, 40000]
})
print(data)
```

Label **Encoding**

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
data['Gender_Label'] = l.fit_transform(data['Gender'])
print(data[['Gender', 'Gender_Label']])
data.head()
```

One-Hot-**Encoding**

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output = False)
city_encoded = ohe.fit_transform(data[['City']])
city_df = pd.DataFrame(city_encoded, columns = ohe.get_feature_names_out(['City']))
print(city_df)
```

Ordinal **Encoding**

```python
from sklearn.preprocessing import OrdinalEncoder
education_order = ['Bachelors', 'Masters', 'PhD']
oe = OrdinalEncoder(categories = [education_order])
data['Education_Ordinal'] = oe.fit_transform(data[['Education']])
print(data[['Education', 'Education_Ordinal']])
```
